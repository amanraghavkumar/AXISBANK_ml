```python
from sklearn.preprocessing import LabelEncoder
Symbol_encoder=LabelEncoder()
Series_encoder=LabelEncoder()
```

```python
data['Symbol']=Symbol_encoder.fit_transform(data['Symbol'])
data['Series']=Series_encoder.fit_transform(data['Series'])
```

```python
data['Date']=pd.to_datetime(data['Date'])
```

```python
data.info()
```

```python
data['Year']=data['Date'].dt.year
data['Month']=data['Date'].dt.month
data['Days']=data['Date'].dt.day
data=data.drop(labels=['Date'],axis=1)
```

```python
data.head()
```

```python
data.info()
```

```python
x=data.drop(labels=['Close'],axis=1)
y=data['Close']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,
                                               random_state=42,
                                               test_size=0.20)
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR=LinearRegression()
LR.fit(x_train,y_train)
```

```python
y_pred=LR.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mse=mean_squared_error(y_pred,y_test)
mae=mean_absolute_error(y_pred,y_test)
score=r2_score(y_pred,y_test)
print('mean squared error',mse)
print('mean absolute error',mae)
print('score',score)
```
